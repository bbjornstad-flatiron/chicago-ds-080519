```python
df = pd.read_csv('titanic.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.drop(columns=['Cabin', 'PassengerId', 'Name'])

df.columns = df.columns.map(lambda s: s.lower())
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df = df.dropna(subset=['embarked'])
df.info()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()

df.age = imputer.fit_transform(df[['age']])
```

```python
df.head()
```

```python
x = df.drop(columns=['survived'])
y = df.survived
```

```python
x_cat = x.select_dtypes(include='object')
x_cat.head()
```

```python
x_cat_dummies = pd.get_dummies(x_cat, drop_first = True)
x_cat_dummies.head()
```

```python
x_num = x.select_dtypes(exclude='object')
x_prep = pd.concat([x_num, x_cat_dummies], axis=1)
x_prep.shape
```

```python
x_prep.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

mms.fit(x_prep)
x_transformed = pd.DataFrame(mms.transform(x_prep), columns=x_prep.columns)
```

```python
x_transformed
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_transformed, y)
```

```python
lrc = LogisticRegression(C=1.0, tol=0.0001)
```

```python
lrc.fit(x_train, y_train)
```

```python
y_hat_test = lrc.predict(x_test)
```

```python
y_hat_train = lrc.predict(x_train)
```

```python
y_test_proba = lrc.predict_proba(x_test)
y_train_proba = lrc.predict_proba(x_train)
```

```python
lrc.score(x_test, y_test)
```

```python
lrc.score(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
```

```python
accuracy_score(y_test, y_hat_test)
```

```python
accuracy_score(y_train, y_hat_train)
```

```python
recall_score(y_test, y_hat_test)
```

```python
recall_score(y_train, y_hat_train)
```

```python
precision_score(y_test, y_hat_test)
```

```python
precision_score(y_train, y_hat_train)
```

```python
f1_score(y_test, y_hat_test)
```

```python
f1_score(y_train, y_hat_train)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_test_proba[:,1])
```

```python
roc_auc_score(y_train, y_train_proba[:,1])
```
