### Step 3. Assign it to a variable called euro12.

```python
# we just use the one in our folder for readability purposes.
euro12 = pd.read_csv('Euro_2012_stats_TEAM.csv')
euro12.head(10)
```

### Step 4. Select only the Goal column.

```python
euro12_goals = euro12.Goals
```

### Step 5. How many team participated in the Euro2012?

```python
euro12.Team.describe()
```

### Step 6. What is the number of columns in the dataset?

```python
euro12.columns.size
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
discipline = euro12.loc[:,['Team', 'Yellow Cards', 'Red Cards']]
discipline.head(10)
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
discipline_sorted = discipline.sort_values(['Red Cards', 'Yellow Cards'], axis=0, ascending=False)
discipline_sorted.head(10)
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
discipline_sorted['Yellow Cards'].describe()['mean']
```

### Step 10. Filter teams that scored more than 6 goals

```python
high_scoring_teams = euro12.loc[euro12['Goals'] >= 6]
high_scoring_teams.head(10)
```
