## Goal 2:  Identify differences between seaborn & matplotlib


### Two code examples to accomplish the same plot:

**Resources:**
- [python graph gallery on seaborn](https://python-graph-gallery.com/seaborn/)
- [seaborn](https://seaborn.pydata.org/)


````python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

# Load in data
tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
````

````python
# Matplotlib:

# Initialize Figure and Axes object
fig, ax = plt.subplots()

# Create violinplot
ax.violinplot(tips["total_bill"], vert=False)

# Show the plot
plt.show()
````

````python
# Seaborn:

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Load the data
tips = sns.load_dataset("tips")

# Create violinplot
sns.violinplot(x = "total_bill", data=tips)

# Show the plot
plt.show()
````

````python
import pandas as pd
df = pd.read_excel('https://query.data.world/s/5qxp2ldwsel3ow2pq5mkvfas2rfaup')
````

````python
df.head()
````

````python
count_ax = sns.countplot(x='state', data=df)
count_ax.tick_params(axis='x', labelrotation=-45)
plt.show()
````

````python
count_ax2 = sns.distplot(a=df['amount'], bins=10)
````

````python
df['amount_scaled'] = df.amount / 1000 # in thousands of dollars instead
violin_ax = sns.violinplot(x='amount_scaled', y='state', data=df)
violin_ax.set_xlabel('Cost in Thousands of Dollars')
violin_ax.set_ylabel('State Where Property is Located')
````
